# Uniform Distribution 均勻分布



注意:連續型  vs 離散型  之差別


wiki:
離散型 https://en.wikipedia.org/wiki/Discrete_uniform_distribution

連續型https://en.wikipedia.org/wiki/Uniform_distribution_(continuous)



numpy.random.uniform是連續型均勻分布

https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.uniform.html#numpy.random.uniform

```
import numpy as np

s = np.random.uniform(0,10,1000)

s
```

```
import matplotlib.pyplot as plt

count, bins, ignored = plt.hist(s, 15, density=True)

# plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')

plt.show()
```
