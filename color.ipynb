# Convert a Photo to Black and White in Python

https://www.blog.pythonlibrary.org/2017/10/11/convert-a-photo-to-black-and-white-in-python/

```
!wget https://www.blog.pythonlibrary.org/wp-content/uploads/2017/10/caterpillar-287x300.jpg
```

```
from PIL import Image
 
def black_and_white(input_image_path,output_image_path):
    color_image = Image.open(input_image_path)
    bw = color_image.convert('L')
    bw.save(output_image_path)

if __name__ == '__main__':  
    black_and_white('caterpillar-287x300.jpg','bw_caterpillar.jpg')
```

```
from PIL import Image
 
def black_and_white_dithering(input_image_path,
    output_image_path,
    dithering=True):
    color_image = Image.open(input_image_path)
    if dithering:
        bw = color_image.convert('1')  
    else:
        bw = color_image.convert('1',
 
    dither=Image.NONE)
    bw.save(output_image_path)


if __name__ == '__main__':
    black_and_white_dithering(
        'caterpillar-287x300.jpg',
        'bw_caterpillar_dithering.jpg')
```
